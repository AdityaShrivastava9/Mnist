# In just 4 lines 

```
from fastai.vision import *
```

```
path = untar_data(URLs.MNIST)
```

```
data = ImageDataBunch.from_folder(path, valid_pct = 0.2, ds_tfms = get_transforms())
```

```
learn = cnn_learner(data, models.resnet50, metrics = accuracy); learn.fit_one_cycle(9)
```

# Results

```
ClassificationInterpretation.from_learner(learn).plot_top_losses(16,  figsize=(15,11))
```
